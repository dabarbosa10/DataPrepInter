```python
def outs(steps):
  outcomes=[]
  for i in range(steps):
    coin=np.random.randint(0,2)
    if coin==0:
      outcomes.append('heads')
    else:
      outcomes.append('tails')
  return outcomes
```

```python
outs(10)
```

```python
def rw_coin(rollings, runs):
  final_tails=[]
  for j in range(runs):
    tails=[0]
    for i in range(rollings):
      coin=np.random.randint(0,2)
      tails.append(tails[i]+coin)
    final_tails.append(tails[-1])

  return final_tails

```

```python
lista1=rw_coin(10, 10000)
```

```python
def rw_dice(lan,rep):
  simu=[]
  for i in range(rep):
    random_walk=[0]
    for i in range(lan):
      step=random_walk[-1]
      dice=np.random.randint(1,7)
      if dice<=2:
        step=max(1,step)
        step=step-1
      elif dice <=5:
        step=step+1
      else:
        step=step+np.random.randint(1,7)
      random_walk.append(step)
    simu.append(random_walk)
  return simu


```

```python
simulation=rw_dice(100,500)
```

```python
aw_np=np.array(simulation)
```

```python
aw_npt=np.transpose(aw_np)
```

```python
plt.plot(aw_npt)
plt.show()
```

```python
ends=aw_npt[-1,:]
plt.hist(ends)
plt.show()
```
